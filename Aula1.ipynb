# Introdução
### 1. Sistema de equações lineares
$$Ax = b$$
*A matriz mxn conhecida
b vetor m conhecido
x vetor n*

* A transformação linear A relaciona o vetor x ao vetor b
* Pode o vetor b ser expresso como combinação linear das colunas de A? Isto é,

$$ b\in [A_{1}, \dots, A_{n}] = \{x_{1}A_{1} + \dots + x_{n}A_{n}\} = \{Ax, x \subset \mathbb{R}^n\} $$

Se sim, o sistema é dito consistente e os coeficientes desta combinação linear são coordenadas do vetor solução de x.

- Quando esta relação existe? 
- Ela é única?
* Vamos considerar sistemas quadrados m = n

### 2. Problemas de autovalor
$$ Ax = \lambda x $$
A: matriz nxn que representa uma transformação linear
$$ \mathbb{R}^n \to \mathbb{R}^n $$
x vetor não nulo de dimensão n
$$ \lambda \in R$$
- O conjunto de todos os autovalores de A é chamado de espectro de A e denotado por $$ \lambda(A) $$ O maior módulo de todos os autovalores de A, $$ max \{|\lambda|, \lambda \in \lambda(A)\} $$
é chamado de raio espectral de A $$ \rho(A) $$

#### Métodos:
Francis (Assimétrica, decomposição QR)
Método da potência / da potência inversa

### 3. Equações não lineares
A forma geral de um sistema de m equações não lineares e n variáveis tem a forma $$ f(x) = 0 
\\f: \mathbb{R}^n \to \mathbb{R}^n \\ x \in \mathbb{R}^n$$ 
tal que as m componentes de f são nulas, chamada raiz ou zero da equação.
#### Exemplos
$$ f(x) = x²-4sen(x)=0 \\ x = 1.93375 \\$$
$$ f(x) = \begin{bmatrix} f(x) \\f_{2}(x) \end{bmatrix} = \begin{bmatrix} x_{1}² - x_{2} + 0.25 \\ -x_{1} +x_{2}² + 0.25 \end{bmatrix} = \begin{bmatrix} 0 \\ 0 \end{bmatrix} $$

$$ \text{Métodos: } m=n=1 $$
Método Bissecção
Método de Newton
Método Secante

$$ \text{Depois: }m=n $$
Método de Newton para Sistemas
